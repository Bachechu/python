```python
import pandas as pd

# 데이터 로드
df = pd.read_csv('iris.csv')

# 1. 결측치 확인하기
print("결측치 확인:")
print(df.isnull().sum())
print(df.isna().sum())

# 2. 중복 데이터 확인하기
print("\n중복 데이터 확인:")
print(df.duplicated().sum())

# 3. 중복 데이터 모두 확인하기
print("\n중복 데이터:")
duplicates = df[df.duplicated()]
print(duplicates)

# 4. 중복 데이터 삭제하기
df_cleaned = df.drop_duplicates()

# 결과 확인
print("\n중복 데이터 삭제 후 데이터프레임 정보:")
print(df_cleaned.info())

# 결측치 처리 (예: 결측치를 0으로 대체)
df_cleaned = df_cleaned.fillna(0)

print("\n결측치 대체 후 확인:")
print(df_cleaned.isnull().sum())
print(df_cleaned.isna().sum())
```

```python
# 1. 품종(species) 열을 기준으로 합계 구하기
species_sum = df.groupby('species').sum()
print("품종별 합계:")
print(species_sum)

# 2. 품종(species) 열을 기준으로 평균 구하기
species_mean = df.groupby('species').mean()
print("\n품종별 평균:")
print(species_mean)
```

```python
import matplotlib.pyplot as plt

# 1. 꽃받침 길이의 시각화
plt.figure(figsize=(10, 6))
plt.bar(df.index, df['sepal_length'], color='blue')
plt.xlabel('Index')
plt.ylabel('Sepal Length')
plt.title('Sepal Length of Iris Dataset')
plt.show()

# 2. 붓꽃 종류별 꽃받침 길이의 평균에 대한 시각화
species_mean_sepal_length = df.groupby('species')['sepal_length'].mean()
species_mean_sepal_length.plot(kind='bar', figsize=(10, 6))
plt.xlabel('Species')
plt.ylabel('Mean Sepal Length')
plt.title('Mean Sepal Length by Species')
plt.show()
```

```python
# 히스토그램 그래프 그리기
df.hist(figsize=(12, 8), bins=20, edgecolor='black')
plt.suptitle('Histogram of Iris Dataset Features', size=16)
plt.show()
```

```python
# 산점도 그래프 그리기
plt.figure(figsize=(10, 6))
plt.scatter(df['sepal_length'], df['petal_length'], c='blue', edgecolor='k', alpha=0.7)
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
plt.title('Scatter Plot of Sepal Length vs Petal Length')
plt.show()
```
